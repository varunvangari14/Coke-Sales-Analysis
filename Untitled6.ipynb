```python
import pandas as pd
import matplotlib.pyplot as plt

rating_col_name = 'Ratings'   # in your file it is 'Ratings'

def map_sentiment(r):
    if pd.isna(r):
        return 'Unknown'
    if r >= 4:
        return 'Positive'
    if r <= 2:
        return 'Negative'
    return 'Neutral'

df['Sentiment'] = df[rating_col_name].apply(map_sentiment)

print("Sentiment counts:")
print(df['Sentiment'].value_counts())
```

```python
sentiment_order = ['Negative', 'Neutral', 'Positive', 'Unknown']
sentiment_counts = (
    df['Sentiment']
    .value_counts()
    .reindex(sentiment_order)
    .fillna(0)
)

plt.figure(figsize=(8, 5))
sentiment_counts.plot(kind='bar')
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import re

negative_reviews = df[(df['Sentiment'] == 'Negative') & df['Review'].notna()]

# Combine all negative review text
all_text = " ".join(negative_reviews['Review'].astype(str).str.lower())

# Keep only words (no numbers, min length 3)
words = re.findall(r'\b[a-z]{3,}\b', all_text)

# Simple stopword list to ignore boring words
stopwords = {
    'the', 'and', 'for', 'that', 'with', 'this', 'you', 'but', 'are', 'was', 'not',
    'have', 'had', 'has', 'from', 'they', 'too', 'just', 'can', 'your', 'out',
    'use', 'using', 'will', 'one', 'like', 'also', 'very', 'more', 'when', 'about',
    'what', 'which', 'how', 'than', 'into', 'then', 'why', 'where', 'who', 'our',
    'their', 'them', 'still', 'could', 'did', 'does', 'didn', 'should', 'would',
    'there', 'here', 'any', 'much', 'even', 'because', 'been', 'only', 'some',
    'really', 'chatgpt', 'gpt', 'ai', 'openai'
}

filtered_words = [w for w in words if w not in stopwords]

word_freq = Counter(filtered_words)
common_issues = word_freq.most_common(30)

print("Top 30 most common words in negative reviews (possible issues):")
for word, freq in common_issues:
    print(f"{word}: {freq}")
```

```python
date_col_name = 'Review Date'  # in your dataset

# Make sure it's datetime (you already did this earlier)
df[date_col_name] = pd.to_datetime(df[date_col_name], errors='coerce')

ts = df.dropna(subset=[date_col_name]).set_index(date_col_name)
```

```python
monthly_sentiment = (
    ts
    .groupby([pd.Grouper(freq='M'), 'Sentiment'])
    .size()
    .unstack(fill_value=0)
)

print("Monthly sentiment counts (last few rows):")
print(monthly_sentiment.tail())
```

```python
plt.figure(figsize=(10, 6))
for sentiment in ['Negative', 'Neutral', 'Positive']:
    if sentiment in monthly_sentiment.columns:
        plt.plot(monthly_sentiment.index,
                 monthly_sentiment[sentiment],
                 marker='o',
                 label=sentiment)

plt.title('Monthly Review Counts by Sentiment')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
monthly_avg_rating = ts[rating_col_name].resample('M').mean()

print("Monthly average rating (last few rows):")
print(monthly_avg_rating.tail())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_rating.index,
         monthly_avg_rating.values,
         marker='o')

plt.title('Monthly Average Rating')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```
